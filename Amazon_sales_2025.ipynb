```python
plt.figure(figsize=(9,4))
ax=sns.countplot(x='Product',data=df,color="red")
ax.bar_label(ax.containers[0])
plt.xticks(rotation=90)
plt.title("Count of Products")
plt.show()
```

```python
plt.figure(figsize=(6,4))
ax=sns.countplot(x='Category',data=df,color="orange")
ax.bar_label(ax.containers[0])

plt.title("Count of Category")
plt.show()
```

```python
plt.figure(figsize=(6,4))
ax=sns.countplot(x='Price',data=df,color="yellow")
ax.bar_label(ax.containers[0])

plt.title("Count of Price")
plt.show()
```

```python
plt.figure(figsize=(6,4))
ax=sns.countplot(x='Customer Location',data=df,color="green")
ax.bar_label(ax.containers[0])
plt.xticks(rotation=90)
plt.title("Count of Location")
plt.show()
```

```python
plt.figure(figsize=(6,4))
ax=sns.countplot(x='Status',data=df,color="green")
ax.bar_label(ax.containers[0])
plt.xticks(rotation=90)
plt.title("Count of Order Status")
plt.show()
```

```python
plt.figure(figsize=(6,4))
ax=sns.countplot(x='Payment Method',data=df,color="c")
ax.bar_label(ax.containers[0])
plt.xticks(rotation=90)
plt.title("Count of Payment Method")
plt.show()
```

```python
plt.figure(figsize=(6,4))
ax=sns.countplot(x='Quantity',data=df,color="brown")
ax.bar_label(ax.containers[0])

plt.title("Count Per Quantity")
plt.show()
```

```python
plt.figure(figsize=(11,4))
sns.countplot(x="Product",data=df,hue="Status")

plt.xticks(rotation=90)
plt.title(" Count of Product By Status")
plt.show()
```

```python

sns.catplot(x = 'Product', y = 'Price', data =df , kind='bar',palette=[ 'blue'])
plt.xticks(rotation=90)
plt.title(" Product With Price")
plt.show()
```

```python
df1 = df.groupby('Product')['Quantity'].sum().reset_index()
```

```python
df1
```

```python
df1 = df1.groupby('Product')['Quantity'].sum().reset_index()

plt.figure(figsize=(8,5))
ax=sns.barplot(data=df1, x='Product', y='Quantity',color='green')
ax.bar_label(ax.containers[0])
plt.title('Product vs Total Quantity')
plt.xticks(rotation=45)
plt.show()
```
